```python
import pandas as pd
```

#Getting the data and cleaning it

Scrapping the data from the URL

```python
url='https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
html = pd.read_html(url, header = 0)
df = html[0]
df.shape
```

```python
df.head()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df.head(10)
```

```python
df.dtypes
```

All type are objects, but most of them must be numeric so we have to convert them

Firstly, I'm going to remove the columns i don't need.

```python
df.drop('Rk', axis = 1, inplace=True)
df.head(5)
```

Age, G and GS should be integers

```python
df['G'].value_counts()
```

```python
df[df['G']=='G'].count()
```

We have 29 rows with no values so we can delete them

```python
df.drop(df[df['G']=='G'].index, axis=0, inplace=True)
```

```python
df['Age']=df['Age'].astype('int64')
df[['G', 'GS']] = df[['G', 'GS']].apply(pd.to_numeric)
df[['MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']] = df[['MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']].apply(pd.to_numeric)
df.dtypes
```

```python
df.isna().sum()
```

```python
df['FG%'].value_counts()
```

There are 28 players with 0 FG%, let's find out it it makes sense.

```python
df[df['FG%']==0.0]
```

It makes sense because they don't have field goals.

```python
df.isna().sum()
```

Let's replace the NaN values for 0

```python
df.fillna(0, inplace=True)
df.isna().sum()
```

```python
df.columns.tolist()
```

The dataset is clean and ready to be analized.

```python
df.describe()
```

What team scores more?

```python
by_team = df.groupby('Tm').sum()
by_team = pd.DataFrame(by_team).reset_index()
by_team
```

```python
by_team[by_team['PTS']==by_team['PTS'].max()]
```

Which is the youngest team?

```python
by_team_avg = df.groupby('Tm').mean()
by_team_avg=pd.DataFrame(by_team_avg).reset_index()
by_team_avg
```

```python
by_team_avg[by_team_avg['Age']==by_team_avg['Age'].min()]['Tm']
```

# Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Top 10 teams points

```python
top10teams = by_team.sort_values(by='PTS', axis=0, ascending=False)[:9]
plt.figure(figsize=(15,10))
plt.bar(top10teams['Tm'],top10teams['PTS'],width = .5)
plt.title('Top 10 teams by points')
plt.ylabel('Total Points')
plt.xlabel('Teams')
plt.show()
```

Best rebounders of the leage

```python
bestrebound = df.sort_values(by='TRB', axis = 0, ascending = False)[:9]
plt.figure(figsize=(15,10))
plt.barh(bestrebound['Player'],bestrebound['TRB'],color='green')
plt.title('Bests Rebounders')
plt.ylabel('Total rebounds')
plt.xlabel('Player')
plt.show()
```

How age is distributed?


```python
plt.figure(figsize=(15,10))
plt.hist(df['Age'], bins=100, width=0.5)
plt.title('Age distribution')
plt.show() 
```

% of points of the Mavericks

```python
dallas = df[df.Tm=='DAL']
dallas = dallas.sort_values(by='PTS', ascending = False).reset_index()

plt.figure(figsize=(15,5))
plt.barh(dallas['Player'][:5], dallas['PTS'][:5])
plt.title('Who scores in Dallas?')
plt.show()
```

```python
dallas_other=dallas[10:].sum()
dallas_other['Player'] = 'Other'
dallas_other = pd.DataFrame(dallas_other[['Player','PTS']]).T
dallastop10= dallas[:10][['Player', 'PTS']]
dallasscores = pd.concat([dallastop10, dallas_other], ignore_index = True)
```

```python
dallasscores
```

```python
plt.figure(figsize=(15,10))
plt.title('%of Points')
plt.pie(dallasscores['PTS'], labels = dallasscores['Player'],autopct='%1.2f%%')

plt.legend()
plt.show()
```
