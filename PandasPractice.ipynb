```python
data.dropna(inplace=True)
```

```python
data.drop('Area code', axis= 1)
```

```python
data['suma'] = data['Total night minutes'] + data['Total eve minutes']
data.head()
```

```python
data['Int Plan binary'] = data['International plan'].apply(lambda x: 1 if x=='Yes' else 0)
data.head(10)
```

```python
choices = ['No', 'Yes']
conditions = [((data['International plan']!='Yes') | (data['Voice mail plan']!='Yes')) , (data['International plan']=='Yes')& (data['Voice mail plan']=='Yes')]
data['Full plan'] = np.select(conditions, choices, default='No')

data.head(20)
```

```python
data['Full plan'].unique().tolist()
```

```python
data['Full plan'].value_counts().tolist()
```

```python
import matplotlib.pyplot as plt

labels = data['Full plan'].unique().tolist()
sizes = data['Full plan'].value_counts().tolist()
explode = (0, 0.2) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
ax1.set_title('% of customers with full plan')

plt.show()
```

```python
labels = data['State'].unique().tolist()
sizes= data['State'].value_counts().tolist()

fig2, ax1 = plt.subplots()
ax1.pie(sizes, labels = labels, shadow = True, frame = True, startangle = 90)
ax1.axis('equal')
ax1.set_title('% of customers per State')
plt.show()
```
