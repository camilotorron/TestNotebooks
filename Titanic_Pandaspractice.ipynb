```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (10,7))
sns.histplot(data=data, x=data['Age'])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.pie(labels = data['Survived'].unique(), x = data['Survived'].value_counts(), autopct='%1.1f%%', explode=[0,0.05], shadow=True, startangle=90)
plt.title('% of Survived')
plt.show()
```

```python
data[['Fare', 'PassengerId']].plot(kind='scatter', x = 'PassengerId', y = 'Fare', figsize=(15,10))
```
