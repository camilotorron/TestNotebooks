# Descriptive analysis

```python
df.describe()
```

Transform 'diagnosis' into numerical 0 and 1

```python
from sklearn.preprocessing import LabelEncoder

lenc = LabelEncoder()
labels = lenc.fit_transform(df['diagnosis'])
df['target'] = labels
df.drop(columns = 'diagnosis', inplace = True, axis = 1)
```

```python
df['target'].value_counts()
```

0 => Benign
1 => Malignant

```python
df.head(5)
```

```python
df.groupby('target').mean()
```

# Visualization

% of Benign/Malignant

```python
import matplotlib.pyplot as plt
```

```python
df.target.value_counts().values
```

```python
plt.figure(figsize=(15,8))
plt.pie(df['target'].value_counts().values,  labels = df['target'].value_counts().index,autopct='%1.2f%%')
plt.legend()
plt.title('% of benign/malignant')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(df.target.value_counts().index,df.target.value_counts().values,width = 1, color=["lightgreen", "cornflowerblue"])
plt.title('Number of Benign/malignant')
plt.ylabel('Cases in dataset')
plt.xlabel('(0 = Benign) (1 = Malignant)')
plt.show()
```

```python
import seaborn as sns

correlation = df.corr()
print(correlation['target'])
```

```python
afectan = correlation['target'].between(-0.3,0.3)
type(afectan)
print(afectan)
```
